## Data Cleaning
Remove duplicates and irrelevant columns.

```python
# Checks columns
df.columns
```

In this price trend analysis, `Product_id` and `Sale` are not useful. Therefore we will remove it from this analysis.

```python
# Keeps only the relevant columns (drop 'Product_id' and 'Sale')
df = df[['weight', 'resoloution', 'ppi',
       'cpu core', 'cpu freq', 'internal mem', 'ram', 'RearCam', 'Front_Cam',
       'battery', 'thickness', 'Price']]
```

```python
df.columns
```

```python
df.duplicated().sum()
```

There are 78 duplicates in the dataset, we will remove them as duplicates can skew statistical analysis.

```python
df.drop_duplicates(inplace = True)
```

```python
df.isna().sum()
```

```python
df.corr()["Price"].sort_values(ascending = False)
```

Strongest **positive** correlation is `ram`, `ppi` and `internal mem`. While strongest **negative** correlation = `thickness`

### Analysis
- 161 entries with no missing values.
- `ram`, `ppi`, and `internal mem` have a strong positive correlation with `Price`. Meaning the higher the RAM/storage, the higher the price.
- `thickness` and `weight` have a strong negative correlation with `Price`. Meaning the thicker and heavier the phone, the lower the price.

## Feature vs Price Visualization
Plot each feature against price to see trends.

```python
sns.heatmap(df.corr(), annot = True, cmap = "rainbow")
```

```python
sns.pairplot(df)
```

```python
colorsforx = ["Red", "Green", "Black", "Purple", "Yellow", "Red", "Green", "Black", "Purple", "Yellow", "Red"]
colorchanger = 0
```

```python
for i in df.columns[:-1]:
    sns.scatterplot(data = df, x = i, y = "Price", color = colorsforx[colorchanger])
    plt.show()
    colorchanger += 1
```

## Final Analysis
Based on the graphs, we can see that the **most influential factors on price** are:
- `ram`, `ppi`, and `internal mem`, where all the three have strong positive correlation. Trends go upward too!
- `thickness` have a strong negative correlation, where the trend go downward.
